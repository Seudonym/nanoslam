```python
import cv2
from visual_odometry import VisualOdometry
import numpy as np
```

```python
video_file = 'mc.mp4'
F = 610
```

```python
cap = cv2.VideoCapture(video_file)
ret, frame = cap.read()
H, W, _ = frame.shape
cap.release()
```

```python
K = np.array(
    [
        [F, 0, W // 2],
        [0, F, H // 2],
        [0, 0, 1],
    ]
)
```

```python
vo = VisualOdometry(K)
```

```python
cap = cv2.VideoCapture(video_file)
```

```python
pose_list = []
```

```python
while cap.isOpened():
    ret, frame = cap.read()
    if not ret: break
    img = cv2.resize(frame, (W, H))
    matches, pose = vo.extract(img)
    pose_list.append(pose)
```
